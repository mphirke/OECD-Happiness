```
happiness = np.array(df['Life satisfaction']).reshape(41,1)
expectancy = np.array(df['Life expectancy']).reshape(41,1)
AP = np.array(df['Air pollution']).reshape(41,1)

print("Air Pollution shape - ", AP.shape)

print(expectancy.shape)
print(expectancy.shape)

f, axs = plt.subplots(1,2,figsize=(20,5))

axs[0].plot(AP, happiness, 'o', color ='b')
axs[0].set_title("Happiness vs Air Pollution")
axs[1].plot(expectancy, happiness, 'x', color='r')
axs[1].set_title("Happiness vs Life Expectancy")
```

```
def gradient_descent(w, x, y, alpha=0.01, n_iters=500, tolerance=0.00001):

    costs = []
    weights = []
    iterations = 0
    print(w.shape)
    print(x.shape)
    print(y.shape)
    
    for i in range(0, n_iters):
        y_estimate = x.dot(w).flatten()
        error = (y.flatten() - y_estimate)
        mse = (1.0/len(x))*np.sum(np.power(error, 2))
        gradient = -(1.0/len(x)) * error.dot(x)

        new_w = w - alpha * gradient


          # Print MSE every 10 iterations
        if iterations % 10 == 0:
            costs.append(mse)
            weights.append(new_w)
            if (iterations % round(n_iters/10)) == 0:
                print("Iteration: %d" % iterations)
                print(mse)
                    

        if np.sum(abs(new_w - w)) < tolerance:
            print("The algorithm has converged.")
            break
        
      
        iterations += 1
        w = new_w

    return new_w, costs, weights
```

```
m = happiness.size
expectancy /= np.max(expectancy)
expectancy_X = np.hstack((np.ones(m).reshape(m, 1), expectancy))

theta = np.ones(2)
theta, costs, weights_exp = gradient_descent(theta, expectancy_X, happiness, alpha=0.01, n_iters=100000)

print("\n\nFinal Weights = ", theta)
```

```
m = happiness.size
AP = AP/ np.max(AP)
AP_X = np.hstack((np.ones(m).reshape(m, 1), AP))
theta_AP = np.ones(2)

theta_AP, costs_AP, weights_AP = gradient_descent(theta_AP, AP_X, happiness, alpha=0.01, n_iters=100000)

print("\n\nFinal Weights = ", theta_AP)
```
